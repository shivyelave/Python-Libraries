# <center>Bar Plot</center>

## 1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt

def main():
    # Load the Titanic dataset from seaborn
    data = sns.load_dataset("titanic")
    
    # Create a bar plot of survival rates by sex
    sns.barplot(data=data, x='sex', y='survived')  # x-axis: 'sex', y-axis: 'survived'
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

# <Center>Pointplot</Center>

## 1. Write a program to draw a point plot for sex against survived for a dataset given in url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv


```python
import seaborn as sns
import matplotlib.pyplot as plt

def main():
    # Load the Titanic dataset from seaborn
    data = sns.load_dataset("titanic")
    
    # Create a point plot of survival rates by sex
    sns.pointplot(data=data, x='sex', y='survived',markers = ['<'])  # x-axis: 'sex', y-axis: 'survived'
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

# <center>Scatter Plot</center>

# <center>Box Plot</center>

## 1. Write a program to draw box plot of life expectancy by continent for a data set given in a url
https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def main():
    # Load the gapminder dataset from seaborn
    data = pd.read_csv('https://raw.githubusercontent.com/resbaz/r-novice-gapminder-files/master/data/gapminder-FiveYearData.csv')

    # Create a box plot
    sns.boxplot(data=data, x='lifeExp', y='continent')  
    plt.title("Box Plot")
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

# <center>Swarm Plot</center>

## 1. Write a program to draw a swarm plot of total bill against size for a given dataset
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv


```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def main():
    # Load the tips dataset from seaborn
    data = sns.load_dataset('tips')
    sns.set(style = 'whitegrid')
    # Create a Swarm plot
    sns.swarmplot(data=data, x='total_bill',y = 'size',native_scale=True)  
    plt.title("Swarm Plot")
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

## 2. Write a program to draw swarm plot of “total bill” against day for a dataset given in url
https://raw.githubusercontent.com/mwaskom/seaborn-data/master/tips.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

def main():
    # Load the tips dataset from seaborn
    data = sns.load_dataset('tips')
    # Create a Swarm plot
    sns.swarmplot(data=data, x='total_bill',y = 'day',hue = 'day')  
    plt.title("Swarm Plot")
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```
