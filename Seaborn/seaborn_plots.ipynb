# <center>Bar Plot</center>

## 1. Write a program to draw bar plot of sex against survived for a dataset given in the url
https://github.com/mwaskom/seaborn-data/blob/master/titanic.csv

```python
import seaborn as sns
import matplotlib.pyplot as plt

def main():
    # Load the Titanic dataset from seaborn
    data = sns.load_dataset("titanic")
    
    # Create a bar plot of survival rates by sex
    sns.barplot(data=data, x='sex', y='survived')  # x-axis: 'sex', y-axis: 'survived'
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```
