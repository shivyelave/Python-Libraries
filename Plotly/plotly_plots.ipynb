```python
# Logging import
from py_logging import (get_debug_logger, get_critical_logger, get_error_logger, get_info_logger, get_warning_logger)
```

## 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as ex
import numpy as np

def main():
    # Generate random x and y coordinates
    X = np.random.rand(1000)
    Y = np.random.rand(1000)

    logger_info = get_info_logger(__name__)
    logger_info.info("X and Y generated.")

    # Create a scatter plot
    fig = ex.scatter(x=X, y=Y ,
                    color=X + Y,
                    color_continuous_scale =  'viridis',
                    title='Scatter Plot of Random Coordinates',
                    size = Y*4,
                    labels={   
                                'x':'X Coordinate'
                                ,'y':'Y Coordinate'})

    # Show the plot
    fig.show()
    logger_info = get_info_logger(__name__)
    logger_info.info("First Program completed.")
if __name__ == '__main__':
    main()
```

## 2. Write a program to draw line and scatter plots for random 100 x and y coordinates

```python
import numpy as np
import plotly.express as ex  # Import Plotly Express for plotting
from py_logging import get_debug_logger  # Import the logging function from your logger_config module

def main():
    # Generate 100 random x and y coordinates between 0 and 1
    X = np.random.random(100)
    Y = np.random.random(100)
    
    # Create a scatter plot with Plotly Express
    # `color` parameter uses the sum of X and Y for coloring the points
    # `color_continuous_scale='viridis'` sets the color scale to 'viridis'
    # `title` sets the title of the scatter plot
    fig = ex.scatter(x=X, y=Y, color=X + Y, color_continuous_scale='viridis', title="Scatter plot")
    fig.show()  # Display the scatter plot
    
    # Create a line plot with Plotly Express
    # `title` sets the title of the line plot
    fig_line = ex.line(x=X, y=Y, title='Line plot')
    fig_line.show()  # Display the line plot
    
    # Get a logger instance configured for INFO level
    logger_debug = get_debug_logger(__name__)
    
    # Log a message indicating the completion of the second part of the program
    logger_debug.debug("Second Program completed.")

# Entry point of the script
if __name__ == '__main__':
    main()
```
