## 1. Write a program to draw a scatter plot for random 1000 x and y coordinates

```python
import plotly.express as ex
import numpy as np

# Generate random x and y coordinates
X = np.random.rand(1000)
Y = np.random.rand(1000)

logger_info = get_info_logger(__name__)
logger_info.info("X and Y generated.")

# Create a scatter plot
fig = ex.scatter(x=X, y=Y ,
                 color=X + Y,
                 color_continuous_scale =  'viridis',
                 title='Scatter Plot of Random Coordinates',
                 size = Y*4,
                 labels={   
                            'x':'X Coordinate'
                            ,'y':'Y Coordinate'})

# Add titles and labels
fig.update_layout(

)

# Show the plot
fig.show()
```
