## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt

def main():
    x = ['Shiv', 'Deven', 'Prayag', 'Ayush']
    y = [3.94, 3.65, 3.75, 4.04]

    plt.plot(x, y, marker='o')  # Added marker='o' to highlight the data points
    plt.xlabel('Names')
    plt.ylabel('Track Score')
    plt.title('Track Scores by Student Name',fontsize = 20)

    plt.show()

if __name__ == '__main__':
    main()
```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:<br>
test.txt<br>
1 2<br>
2 4


```python
import matplotlib.pyplot as plt

def read_data_from_file(filename):

    """
    
    Description:
        Reads x and y axis values from a text file.
    
    Parameters:
        filename: str, the name of the file to read
    
    Returns:
        x: list of floats, x-axis values
        y: list of floats, y-axis values

    """

    x = []
    y = []
    
    with open(filename, 'r') as file:
        for line in file:
            values = line.split()
            x.append(float(values[0]))
            y.append(float(values[1]))
    
    return x, y

def main():
    # Read x and y axis values from the text file
    x, y = read_data_from_file('test.txt')
    
    # Plot the data
    plt.plot(x, y, marker='o', linestyle='-') 
    
    # Label the x-axis
    plt.xlabel('X-axis')
    
    # Label the y-axis
    plt.ylabel('Y-axis')
    
    # Set the title of the plot
    plt.title('Line Graph from File Data')
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    # Execute the main function if this script is run directly
    main()
```

## 4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):<br>
Date,Open,High,Low,Close<br>
03-10-16,774.25,776.065002,769.5,772.559998<br>
04-10-16,776.030029,778.710022,772.890015,776.429993<br>
05-10-16,779.309998,782.070007,775.650024,776.469971<br>
06-10-16,779,780.47998,775.539978,776.859985<br>
07-10-16,779.659973,779.659973,770.75,775.080017

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_graph(x, y):
    """
    Description:
        Plots a graph with the given x and y values.
    
    Parameter:
        x: list-like, x-axis values
        y: list-like, y-axis values
    """
    plt.xlabel('Date')  # Label for the x-axis
    plt.ylabel('Price')  # Label for the y-axis
    plt.title('Prices Over Time')  # Title of the plot
    plt.plot(x, y, marker='o')  # Plot x versus y with markers
    plt.show()  # Display the plot

def main():
    """
    Main function to read data from a CSV file and plot graphs for each price column.
    """
    # Read data from CSV file
    df = pd.read_csv('fdata.csv')
    
    # Define the y-axis values (all columns except the first one)
    y_axis = df.columns[1:]
    
    # Loop through each price column and plot the data
    for price in y_axis:
        x = df['Date']  # x-axis values (dates)
        y = df[price]  # y-axis values (price)
        plot_graph(x, y)  # Call the function to plot the graph

if __name__ == '__main__':
    main()
```

## 5.Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Read data from CSV file
    df = pd.read_csv('fdata.csv')
    
    # Extract columns for plotting
    open_prices = df['Open']   # Open prices
    high_prices = df['High']   # High prices
    low_prices = df['Low']     # Low prices
    close_prices = df['Close'] # Close prices
    
    # Plot each line with different labels
    plt.plot(df['Date'], open_prices, label="Open")   # Plot Open prices
    plt.plot(df['Date'], high_prices, label="High")   # Plot High prices
    plt.plot(df['Date'], low_prices, label="Low")     # Plot Low prices
    plt.plot(df['Date'], close_prices, label="Close") # Plot Close prices
    
    # Label the x-axis
    plt.xlabel('Date')  # Label for the x-axis
    
    # Label the y-axis
    plt.ylabel('Price')  # Label for the y-axis
    
    # Set the title of the plot
    plt.title('Prices Over Time')  # Title of the plot
    
    # Add a legend to the plot
    plt.legend()
    
    # Display the plot
    plt.show()  # Display the plot

if __name__ == '__main__':
    main()
```

## 9. Write a Python program to display the current axis limits values and set new axis values.

```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Read data from CSV file
    df = pd.read_csv('fdata.csv')
    
    # Extract columns for plotting
    open_prices = df['Open']   # Open prices
    high_prices = df['High']   # High prices
    low_prices = df['Low']     # Low prices
    close_prices = df['Close'] # Close prices
    
    # Plot each line with different labels
    plt.plot(df['Date'], open_prices, label="Open", lw =1,ls = '--',c='red',marker = "*")   # Plot Open prices
    plt.plot(df['Date'], high_prices, label="High",lw =2,ls=':',c='blue')   # Plot High prices
    plt.plot(df['Date'], low_prices, label="Low",lw =3,ls='-.',c='black')     # Plot Low prices
    plt.plot(df['Date'], close_prices, label="Close",lw =4,ls = '-') # Plot Close prices
    
    # Label the x-axis
    plt.xlabel('Date')  # Label for the x-axis
    
    # Label the y-axis
    plt.ylabel('Price')  # Label for the y-axis
    
    # Set the title of the plot
    plt.title('Prices Over Time')  # Title of the plot
    
    # Add a legend to the plot
    plt.legend()

    print("Current X limit:",plt.xlim())
    print("Current y limit:",plt.ylim())
    print("Updated x limit:",plt.xlim(-0.2,5))
    print("Updated y limit:",plt.ylim(765,785))
    
    # Display the plot
    plt.show()  # Display the plot

if __name__ == '__main__':
    main()
```
