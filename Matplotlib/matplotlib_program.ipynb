## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:<br>
test.txt<br>
1 2<br>
2 4


```python
import matplotlib.pyplot as plt

def read_data_from_file(filename):

    """
    
    Description:
        Reads x and y axis values from a text file.
    
    Parameters:
        filename: str, the name of the file to read
    
    Returns:
        x: list of floats, x-axis values
        y: list of floats, y-axis values

    """

    x = []
    y = []
    
    with open(filename, 'r') as file:
        for line in file:
            values = line.split()
            x.append(float(values[0]))
            y.append(float(values[1]))
    
    return x, y

def main():
    # Read x and y axis values from the text file
    x, y = read_data_from_file('test.txt')
    
    # Plot the data
    plt.plot(x, y, marker='o', linestyle='-') 
    
    # Label the x-axis
    plt.xlabel('X-axis')
    
    # Label the y-axis
    plt.ylabel('Y-axis')
    
    # Set the title of the plot
    plt.title('Line Graph from File Data')
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    # Execute the main function if this script is run directly
    main()
```

## 4. Write a Python program to draw line charts of the financial data of Alphabet Inc. between October 3, 2016 to October 7, 2016.
Sample Financial data (fdata.csv):<br>
Date,Open,High,Low,Close<br>
03-10-16,774.25,776.065002,769.5,772.559998<br>
04-10-16,776.030029,778.710022,772.890015,776.429993<br>
05-10-16,779.309998,782.070007,775.650024,776.469971<br>
06-10-16,779,780.47998,775.539978,776.859985<br>
07-10-16,779.659973,779.659973,770.75,775.080017

```python
import pandas as pd
import matplotlib.pyplot as plt

def plot_graph(x, y):
    """
    Description:
        Plots a graph with the given x and y values.
    
    Parameter:
        x: list-like, x-axis values
        y: list-like, y-axis values
    """
    plt.xlabel('Date')  # Label for the x-axis
    plt.ylabel('Price')  # Label for the y-axis
    plt.title('Prices Over Time')  # Title of the plot
    plt.plot(x, y, marker='o')  # Plot x versus y with markers
    plt.show()  # Display the plot

def main():
    """
    Main function to read data from a CSV file and plot graphs for each price column.
    """
    # Read data from CSV file
    df = pd.read_csv('fdata.csv')
    
    # Define the y-axis values (all columns except the first one)
    y_axis = df.columns[1:]
    
    # Loop through each price column and plot the data
    for price in y_axis:
        x = df['Date']  # x-axis values (dates)
        y = df[price]  # y-axis values (price)
        plot_graph(x, y)  # Call the function to plot the graph

if __name__ == '__main__':
    main()
```

## 5.Write a Python program to plot two or more lines on same plot with suitable legends of each line.


```python
import pandas as pd
import matplotlib.pyplot as plt

def main():
    # Read data from CSV file
    df = pd.read_csv('fdata.csv')
    
    # Extract columns for plotting
    open_prices = df['Open']   # Open prices
    high_prices = df['High']   # High prices
    low_prices = df['Low']     # Low prices
    close_prices = df['Close'] # Close prices
    
    # Plot each line with different labels
    plt.plot(df['Date'], open_prices, label="Open")   # Plot Open prices
    plt.plot(df['Date'], high_prices, label="High")   # Plot High prices
    plt.plot(df['Date'], low_prices, label="Low")     # Plot Low prices
    plt.plot(df['Date'], close_prices, label="Close") # Plot Close prices
    
    # Label the x-axis
    plt.xlabel('Date')  # Label for the x-axis
    
    # Label the y-axis
    plt.ylabel('Price')  # Label for the y-axis
    
    # Set the title of the plot
    plt.title('Prices Over Time')  # Title of the plot
    
    # Add a legend to the plot
    plt.legend()
    
    # Display the plot
    plt.show()  # Display the plot

if __name__ == '__main__':
    main()
```

## 11. Write a Python program to plot several lines with different format styles in one command using arrays.


```python
import matplotlib.pyplot as plt

def main():
    # Data for plotting
    x = [1, 2, 3, 4]
    y1 = [1, 2, 3, 4]          
    y2 = [4, 5, 6, 7]                
    y3 = [8, 9, 10, 11]         
    y4 = [12, 13, 14, 15]         
    
    # Formats for each line plot
    formats = ['r--', 'g-.', ':', 'k-']  
    
    # List of y-values to plot
    ys = [y1, y2, y3, y4]
    
    # Plot each line with its corresponding format
    for y, format in zip(ys, formats):
        plt.plot(x, y, format)
    
    # Set the title of the plot
    plt.title('Multiple Lines with Different Formats Using Array')
    # Label the x-axis
    plt.xlabel('X-axis')
    # Label the y-axis
    plt.ylabel('Y-axis')
    
    # Add a legend with labels
    plt.legend(['1st', '2nd', '3rd', '4th'], loc='upper right')
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

## 12. Write a Python program to create multiple types of charts

```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    x = np.arange(1, 6)
    y = np.array([2, 3, 5, 7, 11])
    categories = ['A', 'B', 'C', 'D', 'E']
    values = np.random.randint(1, 10, size=5)
    plt.show()
    
    # Line chart
    plt.plot(x, y, 'r--', marker='o', label='Line Chart')
    plt.title('Line Chart')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.show()
    
    # Bar chart
    plt.bar(categories, values, color='skyblue', edgecolor='black')
    plt.title('Bar Chart')
    plt.xlabel('Categories')
    plt.ylabel('Values')
    plt.show()
    
    # Scatter plot
    plt.scatter(x, y, color='green', marker='x')
    plt.title('Scatter Plot')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.show()
    
    # Show all plots
    plt.show()

if __name__ == '__main__':
    main()
```

## 13. Write a Python program to display the grid and draw line charts of the closing value of Alphabet Inc. between October 3, 2016 to October 7, 2016 Customized the grid lines with linestyle -, width .5. and color blue.
Date,Close<br>
03-10-16,772.559998<br>
04-10-16,776.429993<br>
05-10-16,776.469971<br>
06-10-16,776.859985<br>
07-10-16,775.080017


```python
import matplotlib.pyplot as plt
import pandas as pd

def main():
    # Sample data
    data = {
        'Date': ['03-10-16', '04-10-16', '05-10-16', '06-10-16', '07-10-16'],
        'Close': [772.559998, 776.429993, 776.469971, 776.859985, 775.080017]
    }
    
    # Create a DataFrame
    df = pd.DataFrame(data)
    
    # Plotting
    plt.figure(figsize=(10, 6))
    plt.plot(df['Date'], df['Close'], marker='o', linestyle='-', color='b', label='Closing Value')
    
    # Customizing grid
    plt.grid(linestyle='-', linewidth=0.4, color='blue')
    
    # Adding title and labels
    plt.title('Closing Value of Alphabet Inc. (Oct 3, 2016 - Oct 7, 2016)')
    plt.xlabel('Date')
    plt.ylabel('Closing Value')
    
    # Adding legend
    plt.legend()
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```
