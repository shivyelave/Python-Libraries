## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt

def main():
    x = ['Shiv', 'Deven', 'Prayag', 'Ayush']
    y = [3.94, 3.65, 3.75, 4.04]

    plt.plot(x, y, marker='o')  # Added marker='o' to highlight the data points
    plt.xlabel('Names')
    plt.ylabel('Track Score')
    plt.title('Track Scores by Student Name',fontsize = 20)

    plt.show()

if __name__ == '__main__':
    main()
```

## 2. Write a Python program to draw a line using given axis values with suitable label in the x axis, y axis and a title


```python
import matplotlib.pyplot as plt

def main():
    # Define the x-axis values
    x = [2, 3, 4, 5, 6, 7, 8, 9]
    
    
    # Plot the data
    plt.plot(x, c='black', linestyle = '--') 
    
    # Label the x-axis
    plt.xlabel('X-axis')
    
    # Label the y-axis
    plt.ylabel('Y-axis')
    
    # Set the title of the plot with a font size of 20
    plt.title('Line graph', fontsize=20)
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    # Execute the main function if this script is run directly
    main()
```

## 3. Write a Python program to draw a line using given axis values taken from a text file, with suitable label in the x axis, y axis and a title.
Test Data:<br>
test.txt<br>
1 2<br>
2 4


```python
import matplotlib.pyplot as plt

def read_data_from_file(filename):

    """
    
    Description:
        Reads x and y axis values from a text file.
    
    Parameters:
        filename: str, the name of the file to read
    
    Returns:
        x: list of floats, x-axis values
        y: list of floats, y-axis values

    """

    x = []
    y = []
    
    with open(filename, 'r') as file:
        for line in file:
            values = line.split()
            x.append(float(values[0]))
            y.append(float(values[1]))
    
    return x, y

def main():
    # Read x and y axis values from the text file
    x, y = read_data_from_file('test.txt')
    
    # Plot the data
    plt.plot(x, y, marker='o', linestyle='-') 
    
    # Label the x-axis
    plt.xlabel('X-axis')
    
    # Label the y-axis
    plt.ylabel('Y-axis')
    
    # Set the title of the plot
    plt.title('Line Graph from File Data')
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    # Execute the main function if this script is run directly
    main()
```
