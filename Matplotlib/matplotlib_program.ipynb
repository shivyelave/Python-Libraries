## 1. Write a Python program to draw a line with suitable label in the x axis, y axis and a title

```python
import matplotlib.pyplot as plt

def main():
    x = ['Shiv', 'Deven', 'Prayag', 'Ayush']
    y = [3.94, 3.65, 3.75, 4.04]

    plt.plot(x, y, marker='o')  # Added marker='o' to highlight the data points
    plt.xlabel('Names')
    plt.ylabel('Track Score')
    plt.title('Track Scores by Student Name',fontsize = 20)

    plt.show()

if __name__ == '__main__':
    main()
```
