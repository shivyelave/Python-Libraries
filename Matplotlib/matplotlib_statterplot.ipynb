## 1. Write a Python program to draw a scatter graph taking a random distribution in X and Y and plotted against each other.


```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Generate random data for x and y
    x = np.random.random(10)  # 10 random values for x-axis
    y = np.random.random(10)  # 10 random values for y-axis
    
    # Create a scatter plot
    plt.scatter(x, y, marker='*')  # Scatter plot with '*' marker
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

## 2. Write a Python program to draw a scatter plot with empty circles taking a random distribution in X and Y and plotted against each other.


```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Generate random data for x and y
    x = np.random.random(10)  # 10 random values for x-axis
    y = np.random.random(10)  # 10 random values for y-axis
    
    # Create a scatter plot
    plt.scatter(x, y,color='black',facecolor = 'none')  # Scatter plot with '*' marker
    
    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```
