## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>


```python
import matplotlib.pyplot as plt

def main():
    # Data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.bar(programming_languages, popularity)

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()

if __name__ == '__main__':
    main()
```

## 7. Write a Python programming to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    # Create positions for each bar
    positions = np.array([2,3,1,5,6,4])

    # Create the bar chart
    plt.bar(positions, popularity)

    # Add labels and titles
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')

    # Customize x-ticks
    plt.xticks(positions, programming_languages)

    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

## 8. Write a Python programming to display a bar chart of the popularity of programming Languages. Select the width of each bar and their positions.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>

## 9. Write a Python programming to display a bar chart of the popularity of programming Languages. Increase bottom margin.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>


```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    # Create positions for each bar
    positions = np.array([2,3,1,5,6,4])

    # Create the bar chart
    plt.bar(positions, popularity)
    plt.subplots_adjust(bottom = .7)

    # Add labels and titles
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')

    # Customize x-ticks
    plt.xticks(positions, programming_languages)

    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```

## 10. Write a Python program to create bar plot of scores by group and gender. Use multiple X values on the same chart for men and women.<br>
Sample Data:<br>
Means (men) = (22, 30, 35, 35, 26)<br>
Means (women) = (25, 32, 30, 35, 29)

```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data

    Means_men = (22, 30, 35, 35, 26)
    Means_women = (25, 32, 30, 35, 29)
    # Create positions for each bar
    position = np.arange(0,len(Means_men)*3,3)
    # Create the bar chart
    plt.bar(position,Means_men,label = 'Men',width = 1)
    plt.bar(position+1,Means_women,label = 'women',width = 1)
    
    # Add labels and titles
    plt.xlabel('Programming Languages') 
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')

    groups = [ 'Group ' + str(group+1) for group in range(len(Means_men))]
    # Customize x-ticks
    plt.xticks(position+0.5,groups)
    # Display the plot
    plt.legend(loc = 'upper right')
    plt.show()

if __name__ == '__main__':
    main()
```
