## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>


```python
import matplotlib.pyplot as plt

def main():
    # Data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.bar(programming_languages, popularity)

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()

if __name__ == '__main__':
    main()
```

## 2. Write a Python programming to display a horizontal bar chart of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7


```python
import matplotlib.pyplot as plt

def main():
    # Data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.barh(programming_languages, popularity)

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()

if __name__ == '__main__':
    main()
```
