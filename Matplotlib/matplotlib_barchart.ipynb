## 1. Write a Python programming to display a bar chart of the popularity of programming Languages.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7<br>


```python
import matplotlib.pyplot as plt

def main():
    # Data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    plt.bar(programming_languages, popularity)

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()

if __name__ == '__main__':
    main()
```

## 5. Write a Python programming to display a bar chart of the popularity of programming Languages. Attach a text label above each bar displaying its popularity (float value).
Sample data:<br>

```python
import matplotlib.pyplot as plt

def main():
    # Data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    container = plt.bar(programming_languages, popularity)
    plt.bar_label(container,popularity)

    # Add titles and labels
    plt.title('Popularity of Programming Languages')
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')

    plt.show()

if __name__ == '__main__':
    main()
```

## 7. Write a Python programming to display a bar chart of the popularity of programming Languages. Specify the position of each bar plot.
Sample data:<br>
Programming languages: Java, Python, PHP, JavaScript, C#, C++<br>
Popularity: 22.2, 17.6, 8.8, 8, 7.7, 6.7

```python
import matplotlib.pyplot as plt
import numpy as np

def main():
    # Sample data
    programming_languages = ['Java', 'Python', 'PHP', 'JavaScript', 'C#', 'C++']
    popularity = [22.2, 17.6, 8.8, 8, 7.7, 6.7]

    # Create positions for each bar
    positions = np.arange(1,len(programming_languages)+1)


    bar = plt.bar(positions, popularity)

    # Add labels and titles
    plt.xlabel('Programming Languages')
    plt.ylabel('Popularity')
    plt.title('Popularity of Programming Languages')
    plt.bar_label(bar,programming_languages)
    # Customize x-ticks
    plt.xticks(positions)

    # Display the plot
    plt.show()

if __name__ == '__main__':
    main()
```
